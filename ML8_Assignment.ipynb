```python
#import all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## to crawl the webpage

```python
import urllib.request
response = urllib.request.urlopen('http://php.net/')
html = response.read()
```

```python
#import the beautiful soup which is very useful utility for web scraping. 

from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
tokens = [t for t in text.split()]
print (tokens)
```

```python
#to know the Word Frequency in data
import nltk

freq = nltk.FreqDist(tokens)
for key,value in freq.items():
    print (str(key) + ':' + str(value))

```

```python
#plotting the frequency of count 
freq.plot(20, cumulative=False)
```

```python
#Remove Stop Words Using NLTK
from nltk.corpus import stopwords
stopwords.words('english')

clean_tokens = tokens[:]
sr = stopwords.words('english')
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
```

```python
clean_tokens
```

```python
#now take the frequency of token/words
frequency = nltk.FreqDist(clean_tokens)
for key,val in frequency.items():
    print (str(key) + ':' + str(val))
```

```python
#plotting the graph
frequency.plot(20, cumulative=False)
```
